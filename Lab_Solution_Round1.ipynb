```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type.
```

```python
# files = files.astype({"Customer Lifetime Value" : float}, errors="ignore")
files["Customer Lifetime Value"]
files.dtypes
```

```python
def multiply(x):
    t = ""
    for char in x:
        if char == ".":
            char ==" "
        else:
            t = t + char 
    return int(100*t) 


# files["Customer Lifetime Value"] = list(map(multiply, files["Customer Lifetime Value"]))
files["Customer Lifetime Value"] = list(map(multiply, files["Customer Lifetime Value"]))
files["Customer Lifetime Value"]
```
